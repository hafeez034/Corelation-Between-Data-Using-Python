## Filling missing numericals using sklearn

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer(strategy="mean")
ar =si.fit_transform(df[['Score', 'votes', 'budget', 'gross', 'runtime']])
ar
```

```python
df2=pd.DataFrame(ar,columns = df.select_dtypes(include='float').columns)
```

```python
df2
```

```python
df2.astype("int64")
```

```python
newdf = df+df2
```

```python
newdf   ## This Method is not joining Data Frames Properly
```

```python
df[df.select_dtypes(include='float').columns] =df2      ## In this Step We have added new non-null values to the original df
```

```python
df[['Score','votes','budget','gross']] = df[['Score','votes','budget','gross']].astype("int64")
```

```python
df.head(3)
```

```python
df.sort_values(by=['gross'],inplace=False,ascending = False)
```

```python
## Now Finding corelation between different parameters
```

```python
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget Vs Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross')
plt.show()
```

```python
##Using Seaborn For Graph
sn.regplot(x='budget',y='gross',data=df,scatter_kws={'color':'orange'} , line_kws={'color':'blue'})
            
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype =='object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] =  df_numerized[col_name].cat.codes
        
df_numerized
```

```python
df_numerized.corr()
```

```python
cr_matrix = df_numerized.corr()
sn.heatmap(cr_matrix,annot=True)
plt.show() 
```

```python
cr_matrix_high = cr_matrix.where(cr_matrix > 0.2)
sn.heatmap(cr_matrix_high , annot=True, cmap='coolwarm', vmin=-1, vmax=1)
```

```python
## From the Above Graph We can conclude that direector budget,gross,writer and director are corelated which will give good score and votes
```

```python
cr_matrix_highest = cr_matrix.where(cr_matrix > 0.3)
sn.heatmap(cr_matrix_highest , annot=True, cmap='coolwarm', vmin=-1, vmax=1)
```

```python
## From the above we can also conclude that budget, gross , director and runtime are corelated
```
